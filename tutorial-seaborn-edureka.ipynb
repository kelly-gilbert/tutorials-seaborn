### Plotting functions

* relplot - plots RELATIONSHIP (scatter or line)
* catplot - plots category (kind = point, bar count, strip, swarm, box, violin, boxen)
* distplot - univariate distribution; shows histogram and kernel density estimate
* jointplot - bivariate distribution (plot with marginal histograms); "scatter" | "reg" (regression) | "resid" | "kde" (heatmap)| "hex" 
* GridPlot - small multiples (charts in a grid)
* PairPlot - matrix of relational charts (each variable vs. each variable), like a splom in R


### relplot (relational)
Show the relationship between two variables.

```python
# scatterplot
sns.relplot(data=a, x='passengers', y='month')
```

```python

# vary the color by year
sns.relplot(data=a, x='passengers', y='month', hue='year')
```

```python

b = sns.load_dataset('tips')
```

```python

# specify a line chart instead of scatter
sns.relplot(data=b, x='time', y='tip', kind='line')
```


### catplot (categorical)
Show the relationship between two variables.

```python
sns.catplot(data=b, x='day', y='total_bill')
```

```python
# change the kind
sns.catplot(data=b, x='day', y='total_bill', kind='swarm')
```

### distplot (distribution)
Univariate distribution

```python

from scipy import stats
```

```python
# randomly sample 100 data points from a normal distribution w/ mean 5 and stdev 2
c = np.random.normal(loc=5, size=100, scale=2)
```

```python
sns.distplot(c)
```


### jointplot
Show the relationship between two variables.

```python
c2 = pd.DataFrame(c, columns=['col1'])
c2['col2'] = np.random.normal(loc=50, size=100, scale=2)
```

```python
c2
```

```python
sns.jointplot(data=c2, x='col1', y='col2')
```

```python
sns.jointplot(data=c2, x='col1', y='col2', kind='hex')
```


### facetplot (multi-plot grid)

```python
# facetplot
a = sns.load_dataset('iris')

# create separate plots by species (in cols)
b = sns.FacetGrid(data=a, col='species')

# use the map function to generate histograms of sepal length
b.map(plt.hist, 'sepal_length')
```


### PairGrid (plots pairwise relationships between variables)

```python

# pairgrid example (this is kind of like a splom in R)
a = sns.load_dataset('flights')
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python

# make some fake data
a = pd.DataFrame({ 'col1' : np.random.normal(loc=0, size=100, scale=2), 
                   'col2' : np.random.normal(loc=10, size=100, scale=2),
                   'col3' : np.random.normal(loc=50, size=100, scale=2)})
a
```

```python

# try with more cols
b = sns.PairGrid(a)
b.map(plt.scatter)
```


### Changing style

```python
# darkgrid is a gray background with white gridlines
sns.set(style='darkgrid')
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python

# dark is gray background, but without the gridlines
sns.set(style='dark')
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python

# white is white bg without gridlines
sns.set(style='white', color_codes=True)
a = sns.load_dataset('tips')
sns.boxplot(data=a, x='day', y='total_bill')
```

```python

# change the axes
sns.set(style='white', color_codes=True)
a = sns.load_dataset('tips')
sns.boxplot(data=a, x='day', y='total_bill')
sns.despine(offset=10, trim=True)    # start the axis at 10 and trim it before that
```
