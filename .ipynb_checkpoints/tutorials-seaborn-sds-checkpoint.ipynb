
### Joint kernel density estimate for bivariate data

```python
sns.set(style='white')

# generate 500 sample values
random_state = np.random.RandomState(10)

mean = [2,2]
cov = [(1, 0.5), (0.5, 1)]
x1, x2 = np.random.multivariate_normal(mean, cov, 500).T

x1 = pd.Series(x1, name='Example_X')
x2 = pd.Series(x2, name='Example_Y')

example_three = sns.jointplot(x1, x2, kind='kde', height=7, space=0, color='g')
    # space is the space between the marginal kde plots and the heatmap
```
